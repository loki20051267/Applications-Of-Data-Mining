```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("titanic.csv")
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df)
plt.title('Count Plot: Survival Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Pclass', y='Fare', data=df)
plt.title('Line Chart: Fare vs Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Survival Proportion')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(df['Age'], df['Fare'], gridsize=30, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Hexbin Plot: Age vs Fare')
plt.show()
```
